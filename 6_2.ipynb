```python
survival_by_sex = df_titanic.groupby(['Survived', 'Pclass']).size().unstack()

survival_by_sex.plot(kind='bar', stacked=True, color=['gray', 'brown', 'olive'])
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Pclass', loc='upper right')

plt.show()
```

Есть ли счастливый билет?
помог ли счастливый билет - выжить?

```python
df_titanic['LuckyTicket'] = df_titanic['Ticket'].apply(lambda x: 1 if sum(int(digit) for digit in str(x) if digit.isdigit()) % 10 == 0 else 0)
```

```python
sns.countplot(x='Survived', hue='LuckyTicket', data=df_titanic)
plt.legend(['Unlucky Ticket', 'Lucky Ticket'])
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival by Lucky Ticket')
plt.show()
```

```python
lucky_ticket_survived = df_titanic[df_titanic['LuckyTicket'] == 1]['Survived'].mean()
unlucky_ticket_survived = df_titanic[df_titanic['LuckyTicket'] == 0]['Survived'].mean()

print("Коэффициент выживаемости со счастливым билетом: {:.2f}%".format(lucky_ticket_survived * 100))
print("Выживаемость без счастливого билета: {:.2f}%".format(unlucky_ticket_survived * 100))
```
