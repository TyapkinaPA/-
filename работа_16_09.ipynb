##Срезы
###Срез (англ. slice) - быстрое получение части списка

```python
a = [7, 9, 6, 6, 9, 1, 5]

# Список[начало : конец]
# Список[начало : конец : шаг]

print(a[3:6])       # [6, 9, 1]
print(a[1:2])       # [9]

print(a[1:6:2])     # [9, 6, 1]

print(a[:4])        # [7, 9, 6, 6]
print(a[5:])        # [1, 5]
print(a[:])         # [7, 9, 6, 6, 9, 1, 5]
```

```python
a[0:2] = [1, 12]            # Замена нескольких элементов
a[0:2] = []                 # Удаление нескольких элементов
a[1:1] = ['asd', 'qwrt']    # Вставка нескольких элементов
a[:0] = a                   # Копия самого себя в начало списка
a[:] = []                   # удаление всех элементов из списка
b = a                       # Создание ссылки на список
b = a[:]                    # Создание копии списка
```

##Списковая сборка
a = [*выражение* for *переменная* in *значение* (if *условие*)]

```python
a = [1*5 for i in 4, 13, 26, 2, 8]
print(a)       # [20, 65, 130, 10, 40]

a = [1*5 for i in 4, 13, 26, 2, 8 if i < 10]
print(a)       # [20, 10, 40]
```
